We need to address the AR(2) model given by:

$$x_t = -0.9 x_{t-2} + w_t$$

This is a second-order autoregressive model. To proceed, we'll follow these steps:

### 1. **Write the autoregressive polynomial**

The AR(2) model can be represented in polynomial form. The general form for an AR(2) process is:

$$x_t = \phi_1 x_{t-1} + \phi_2 x_{t-2} + w_t$$

For our case, the model is:

$$x_t = -0.9 x_{t-2} + w_t$$

Thus, $\phi_1 = 0$ and $\phi_2 = -0.9$.

The autoregressive polynomial is defined as:

$$\phi(B) = 1 - \phi_1 B - \phi_2 B^2$$

Substituting the values of $\phi_1$ and $\phi_2$:

$$\phi(B) = 1 + 0.9B^2$$

### 2. **Find the roots of the autoregressive polynomial**

The roots of the AR polynomial $\phi(B) = 1 + 0.9 B^2 = 0$ are found by solving:

$$1 + 0.9 B^2 = 0$$

Thus,

$$B^2 = -\frac{1}{0.9} = -1.111$$

The roots are:

$$B = \pm \sqrt{-1.111} = \pm i \cdot \sqrt{1.111} = \pm 1.054i$$

These are complex conjugate roots, indicating a process with oscillatory behavior.

### 3. **Plot the Autocorrelation Function (ACF)**

The next step is to plot the autocorrelation function $\rho(h)$ of this AR(2) model. For a second-order autoregressive process, the ACF decays with oscillations since the roots are complex. We can use R to simulate this process and plot its ACF.

#### R Code Explanation:

- **Simulation of AR(2)**: We'll simulate a time series based on the AR(2) model.
- **ACF plot**: We'll then calculate and plot the autocorrelation function of the simulated data.

#### R Code:
```R
# Set seed for reproducibility
set.seed(123)

# Define AR(2) model parameters
ar_params <- c(0, -0.9)

# Simulate an AR(2) process with 100 observations
simulated_data <- arima.sim(n = 100, list(ar = ar_params))

# Plot the simulated data
plot(simulated_data, type = "l", main = "Simulated AR(2) Process", ylab = "x_t", xlab = "Time")

# Plot the ACF
acf(simulated_data, main = "ACF of the AR(2) Process")
```

#### Explanation:
1. **Set seed**: To ensure the results are reproducible, we set a seed.
2. **Define AR(2) parameters**: `ar_params` represents the AR coefficients ($\phi_1 = 0$, $\phi_2 = -0.9$).
3. **Simulate the AR(2) process**: We use the `arima.sim()` function to simulate 100 observations of the AR(2) process.
4. **Plot the ACF**: The `acf()` function computes and plots the autocorrelation function.

### 4. **Interpretation of ACF**

Given that the roots of the characteristic equation are complex, we expect the ACF to show an **oscillating** pattern with a **gradual decay**. The ACF of an AR(2) process with complex roots typically exhibits damped sinusoidal behavior.

### Summary of Results:

- The roots of the autoregressive polynomial are complex ($\pm 1.054i$).
- The ACF of the process will oscillate and decay as expected for an AR(2) process with complex conjugate roots.

By running the above R code, you will see the oscillatory decay in the ACF, which is characteristic of an AR(2) model with complex roots.

### 1. **Viết đa thức tự hồi quy**

$$x_t = \phi_1 x_{t-1} + \phi_2 x_{t-2} + w_t$$

Với mô hình được cho:

$$x_t = -0.9 x_{t-2} + w_t$$

Ở đây, \(\phi_1 = 0\) và \(\phi_2 = -0.9\).

Do đó, đa thức tự hồi quy của mô hình sẽ là:

$$\phi(B) = 1 - \phi_1 B - \phi_2 B^2$$

Thay giá trị của \(\phi_1\) và \(\phi_2\) vào:

$$\phi(B) = 1 + 0.9B^2$$

### 2. **Tìm nghiệm của đa thức tự hồi quy**

Chúng ta cần tìm nghiệm của phương trình \(\phi(B) = 1 + 0.9 B^2 = 0\), tức là:

$$1 + 0.9 B^2 = 0$$

Kết quả là:

$$B^2 = -\frac{1}{0.9} = -1.111$$

Nghiệm sẽ là:

$$B = \pm \sqrt{-1.111} = \pm i \cdot \sqrt{1.111} = \pm 1.054i$$

Như vậy, chúng ta có nghiệm phức liên hợp, biểu hiện rằng quá trình này có tính dao động.

### 3. **Vẽ biểu đồ hàm tự tương quan (ACF)**

Bước tiếp theo là vẽ đồ thị hàm tự tương quan (ACF) của mô hình AR(2). Với một mô hình tự hồi quy bậc 2 (AR(2)) có nghiệm phức, ACF sẽ dao động và suy giảm theo thời gian. Chúng ta có thể sử dụng R để mô phỏng quá trình này và vẽ ACF.

#### Giải thích mã R:

- **Mô phỏng AR(2)**: Chúng ta sẽ mô phỏng một chuỗi thời gian dựa trên mô hình AR(2).
- **Vẽ ACF**: Sau đó, chúng ta tính toán và vẽ đồ thị ACF của dữ liệu đã mô phỏng.

#### Mã R:
```R
# Đặt seed để tái hiện kết quả
set.seed(123)

# Định nghĩa tham số AR(2)
ar_params <- c(0, -0.9)

```

#### Giải thích từng bước:

2. **Định nghĩa tham số AR(2)**: `ar_params` đại diện cho các hệ số AR (\(\phi_1 = 0\), \(\phi_2 = -0.9\)).
3. **Mô phỏng quá trình AR(2)**: Sử dụng hàm `arima.sim()` để mô phỏng 100 quan sát của quá trình AR(2).
4. **Vẽ ACF**: Hàm `acf()` tính toán và vẽ hàm tự tương quan.

### 4. **Giải thích ACF**

Vì nghiệm của đa thức đặc trưng là phức, chúng ta dự đoán ACF sẽ có dạng **dao động** và **giảm dần**. Đồ thị ACF của một quá trình AR(2) với nghiệm phức thường thể hiện tính chất dao động tắt dần theo thời gian.
