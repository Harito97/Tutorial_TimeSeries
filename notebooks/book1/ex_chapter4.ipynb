# 4.1

**Quy Trình Đã Cho:**

$$Y_t = 5 + e_t - (1/2)e_{t-1} + (1/4)e_{t-2}$$

**Tính Toán Phương Sai:**

$$
Var(Y_t) = Var(5) + Var(e_t) + (1/4)Var(e_{t-1}) + (1/16)Var(e_{t-2})
         = 0 + \sigma_e^2 + (1/4)\sigma_e^2 + (1/16)\sigma_e^2  
         = (21/16)\sigma_e^2
$$

**Tính Toán Tự Hiệp Phương:**

**Độ Trễ 1:**  

$$
Cov(Y_t, Y_{t-1}) = Cov(e_t, -1/2 e_{t-1}) + Cov(-1/2 e_{t-1}, e_{t-1})  
                    = -1/2 Cov(e_t, e_{t-1}) - 1/2 Var(e_{t-1})  
                    = -1/2 * 0 - 1/2 * \sigma_e^2  
                    = -1/2 \sigma_e^2
$$

**Độ Trễ 2:**  

$$
Cov(Y_t, Y_{t-2}) = Cov(e_t, 1/4 e_{t-2}) + Cov(-1/2 e_{t-1}, 1/4 e_{t-2}) + Cov(1/4 e_{t-2}, e_{t-2})  
                    = 1/4 Cov(e_t, e_{t-2}) - 1/8 Cov(e_{t-1}, e_{t-2}) + 1/4 Var(e_{t-2})  
                    = 1/4 * 0 - 1/8 * 0 + 1/4 * \sigma_e^2  
                    = 1/4 \sigma_e^2
$$

**Độ Trễ 3:**  
$$
Cov(Y_t, Y_{t-3}) = 0
$$

**Tự Tương Quan:**  

$$
ρ_k = Cov(Y_t, Y_{t-k}) / Var(Y_t)
$$
$$
ρ_0 = 1  
$$
$$
ρ_1 = (-1/2 \sigma_e^2) / ((21/16)\sigma_e^2) = -10/21  
$$
$$
ρ_2 = (1/4 \sigma_e^2) / ((21/16)\sigma_e^2) = 4/21  
$$
$$
ρ_3 = 0
$$

**Kết Luận:**  
Quy trình thời gian đã cho là một quy trình MA(2) (Trung Bình Di Động bậc 2) với các tự tương quan sau:  

$$
ρ_0 = 1  
$$
$$
ρ_1 = -10/21  
$$
$$
ρ_2 = 4/21  
$$
$$
ρ_k = 0 | k > 2  
$$

Điều này cho thấy giá trị hiện tại của quy trình bị ảnh hưởng bởi thành phần nhiễu trắng hiện tại và hai thành phần nhiễu trắng trước đó. Hàm tự tương quan cắt đứt sau độ trễ 2, điều này đặc trưng cho một quy trình MA(2).



# 4.4

Cach khac la thay truc tiep cong thuc vao:
$$\rho_{1} = \frac{}{}

Để chứng minh rằng khi $\theta$ được thay bằng $\frac{1}{\theta}$, hàm tự tương quan (autocorrelation function - ACF) của quá trình MA(1) không thay đổi, ta cần xét quá trình MA(1) với hai trường hợp khác nhau.

### 1. Quá trình MA(1) ban đầu
Giả sử chuỗi thời gian $\{Y_t\}$ là một quá trình MA(1) được định nghĩa bởi:

$$
Y_t = \epsilon_t + \theta \epsilon_{t-1}
$$

trong đó:
- $\{\epsilon_t\}$ là nhiễu trắng với trung bình bằng 0 và phương sai là $\sigma^2$: $E(\epsilon_t) = 0$, $Var(\epsilon_t) = \sigma^2$,
- $\theta$ là tham số của mô hình.

### 2. Hàm tự tương quan của quá trình MA(1)
- **Trung bình**: Do $\epsilon_t$ có trung bình bằng 0, nên:

$$
E(Y_t) = E(\epsilon_t + \theta \epsilon_{t-1}) = 0
$$

- **Phương sai**: Phương sai của $Y_t$ là:

$$
Var(Y_t) = Var(\epsilon_t + \theta \epsilon_{t-1}) = \sigma^2 (1 + \theta^2)
$$

- **Tự hiệp phương sai tại lag 0** ($\gamma(0)$):

$$
\gamma(0) = Cov(Y_t, Y_t) = Var(Y_t) = \sigma^2 (1 + \theta^2)
$$

- **Tự hiệp phương sai tại lag 1** ($\gamma(1)$):

$$
\gamma(1) = Cov(Y_t, Y_{t-1}) = E[(\epsilon_t + \theta \epsilon_{t-1})(\epsilon_{t-1} + \theta \epsilon_{t-2})]
$$

Do $E(\epsilon_t \epsilon_{t-1}) = 0$ khi $t \neq t-1$ và $E(\epsilon_t^2) = \sigma^2$, ta có:

$$
\gamma(1) = E(\epsilon_t \epsilon_{t-1}) + \theta E(\epsilon_{t-1}^2) = \theta \sigma^2
$$

- **Tự hiệp phương sai tại lag 2 trở đi** ($\gamma(k) = 0$ với $k \geq 2$):

$$
\gamma(k) = 0 \text{ khi } k \geq 2
$$

- **Hàm tự tương quan (ACF)**: Hàm tự tương quan $\rho(k)$ là:

$$
\rho(k) = \frac{\gamma(k)}{\gamma(0)}
$$

Với $k = 1$:

$$
\rho(1) = \frac{\gamma(1)}{\gamma(0)} = \frac{\theta \sigma^2}{\sigma^2 (1 + \theta^2)} = \frac{\theta}{1 + \theta^2}
$$

Với $k \geq 2$:

$$
\rho(k) = 0 \text{ khi } k \geq 2
$$

### 3. Khi $\theta$ được thay bằng $\frac{1}{\theta}$
Xét quá trình MA(1) mới khi $\theta$ được thay bằng $\frac{1}{\theta}$:

$$
Y_t = \epsilon_t + \frac{1}{\theta} \epsilon_{t-1}
$$

Ta tính lại hàm tự tương quan $\rho(k)$ trong trường hợp này:

- **Phương sai**: Phương sai của $Y_t$ bây giờ là:

$$
Var(Y_t) = \sigma^2 \left( 1 + \left(\frac{1}{\theta}\right)^2 \right) = \sigma^2 \frac{1 + \theta^2}{\theta^2}
$$

- **Tự hiệp phương sai tại lag 1**:

$$
\gamma(1) = \frac{1}{\theta} \sigma^2
$$

- **Hàm tự tương quan tại lag 1**:

$$
\rho(1) = \frac{\gamma(1)}{\gamma(0)} = \frac{\frac{1}{\theta} \sigma^2}{\sigma^2 \frac{1 + \theta^2}{\theta^2}} = \frac{\theta}{1 + \theta^2}
$$

Vậy hàm tự tương quan $\rho(1)$ vẫn giữ nguyên giá trị là $\frac{\theta}{1 + \theta^2}$ khi thay $\theta$ bằng $\frac{1}{\theta}$. Điều này chứng minh rằng hàm tự tương quan của quá trình MA(1) không thay đổi khi $\theta$ được thay bằng $\frac{1}{\theta}$.

### Kết luận:
Hàm tự tương quan của quá trình MA(1) không thay đổi khi thay $\theta$ bằng $\frac{1}{\theta}$.

# 4.7

Hàm tự tương quan (autocorrelation function - ACF) là một công cụ quan trọng trong việc phân tích các mô hình chuỗi thời gian, đặc biệt là các mô hình trung bình động (Moving Average - MA). Dưới đây là các đặc điểm quan trọng của hàm tự tương quan cho hai mô hình MA phổ biến: MA(1) và MA(2).

### (a) Mô hình MA(1)
Mô hình MA(1) được định nghĩa như sau:

$$
Y_t = \epsilon_t + \theta \epsilon_{t-1}
$$

trong đó:
- $\epsilon_t$ là nhiễu trắng (white noise) với trung bình bằng 0 và phương sai $\sigma^2$.
- $\theta$ là tham số của mô hình.

**Đặc điểm của hàm tự tương quan (ACF) của mô hình MA(1):**
- **Trung bình**: Vì $\epsilon_t$ có trung bình bằng 0, nên $Y_t$ cũng có trung bình bằng 0: $E(Y_t) = 0$.
- **Phương sai**: Phương sai của $Y_t$ là:
  
  $$
  Var(Y_t) = \sigma^2 (1 + \theta^2)
  $$

- **Hàm tự tương quan tại lag 0**: Đây là phương sai của $Y_t$, tức là:

  $$
  \gamma(0) = \sigma^2 (1 + \theta^2)
  $$

- **Hàm tự tương quan tại lag 1**: ACF tại lag 1 là:

  $$
  \gamma(1) = \theta \sigma^2
  $$

  Tự tương quan chuẩn hóa (ACF tại lag 1) là:

  $$
  \rho(1) = \frac{\gamma(1)}{\gamma(0)} = \frac{\theta}{1 + \theta^2}
  $$

- **Hàm tự tương quan tại các lag lớn hơn 1**: Với mọi lag $k \geq 2$, hàm tự tương quan là 0:

  $$
  \rho(k) = 0 \text{ khi } k \geq 2
  $$

**Kết luận cho mô hình MA(1)**: ACF của MA(1) giảm nhanh sau lag 1 và trở thành 0 cho các giá trị lag lớn hơn 1. Điều này có nghĩa là mô hình MA(1) có ký ức ngắn hạn, với tương quan chỉ tồn tại trong 1 bước thời gian.

---

### (b) Mô hình MA(2)
Mô hình MA(2) được định nghĩa như sau:

$$
Y_t = \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2}
$$

trong đó:
- $\epsilon_t$ là nhiễu trắng với trung bình bằng 0 và phương sai $\sigma^2$.
- $\theta_1$ và $\theta_2$ là các tham số của mô hình.

**Đặc điểm của hàm tự tương quan (ACF) của mô hình MA(2):**
- **Trung bình**: Tương tự như mô hình MA(1), $E(Y_t) = 0$ vì trung bình của nhiễu trắng $\epsilon_t$ bằng 0.
- **Phương sai**: Phương sai của $Y_t$ là:

  $$
  Var(Y_t) = \sigma^2 (1 + \theta_1^2 + \theta_2^2)
  $$

- **Hàm tự tương quan tại lag 1**: ACF tại lag 1 là:

  $$
  \gamma(1) = \theta_1 \sigma^2 + \theta_1 \theta_2 \sigma^2
  $$

  Tự tương quan chuẩn hóa (ACF tại lag 1):

  $$
  \rho(1) = \frac{\gamma(1)}{\gamma(0)} = \frac{\theta_1 + \theta_1 \theta_2}{1 + \theta_1^2 + \theta_2^2}
  $$

- **Hàm tự tương quan tại lag 2**: ACF tại lag 2 là:

  $$
  \gamma(2) = \theta_2 \sigma^2
  $$

  Tự tương quan chuẩn hóa (ACF tại lag 2):

  $$
  \rho(2) = \frac{\gamma(2)}{\gamma(0)} = \frac{\theta_2}{1 + \theta_1^2 + \theta_2^2}
  $$

- **Hàm tự tương quan tại các lag lớn hơn 2**: Với mọi lag $k \geq 3$, hàm tự tương quan là 0:

  $$
  \rho(k) = 0 \text{ khi } k \geq 3
  $$

**Kết luận cho mô hình MA(2)**: ACF của MA(2) sẽ có giá trị khác 0 ở hai lag đầu tiên (lag 1 và lag 2), nhưng sau đó trở thành 0 cho các giá trị lag lớn hơn 2. Điều này cho thấy mô hình MA(2) có ký ức dài hơn một chút so với MA(1), nhưng vẫn là một quá trình trung bình động ngắn hạn.
