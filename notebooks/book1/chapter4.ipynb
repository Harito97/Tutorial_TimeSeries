# Các mô hình cho chuỗi thời gian dừng

**Tóm tắt Nội dung Chính Chương 4 và Mục đích của các Mô hình Phân tích Chuỗi Thời gian**

Chương 4 của cuốn sách tập trung vào các **mô hình chuỗi thời gian tuyến tính**, là nền tảng cho việc phân tích chuỗi thời gian. Dưới đây là nội dung chính và mục đích của các mô hình được đề cập:

* **Mô hình trung bình di động (MA)**: Mô hình này **biểu diễn một giá trị trong chuỗi thời gian là một hàm tuyến tính của các giá trị nhiễu trắng hiện tại và trong quá khứ**. Mục đích của mô hình MA là nắm bắt **tác động ngắn hạn** của các cú sốc ngẫu nhiên lên chuỗi thời gian. 

   * **Đặc điểm**: Các mô hình MA có **hàm tự tương quan** (ACF) bị cắt sau độ trễ q (bậc của mô hình), trong khi **hàm tự tương quan riêng phần** (PACF) giảm dần.
   * **Ứng dụng**: Mô hình MA phù hợp để phân tích chuỗi thời gian có các mẫu **biến động ngắn hạn**, chẳng hạn như dữ liệu tài chính hoặc dữ liệu thời tiết. 

* **Mô hình tự hồi quy (AR)**: Mô hình AR biểu diễn một giá trị trong chuỗi thời gian là một hàm tuyến tính của các giá trị **trong quá khứ** của nó và một thuật ngữ nhiễu trắng. Mục đích của mô hình AR là nắm bắt **sự phụ thuộc** giữa các quan sát trong chuỗi thời gian.

   * **Đặc điểm**:  Mô hình AR bậc p có PACF bị cắt sau độ trễ p, trong khi ACF giảm dần.
   * **Ứng dụng**: Mô hình AR thích hợp để phân tích các chuỗi thời gian thể hiện **xu hướng** hoặc **chu kỳ**, ví dụ như dữ liệu kinh tế vĩ mô. 

* **Mô hình tự hồi quy trung bình di động (ARMA)**: Mô hình ARMA là sự kết hợp của mô hình AR và MA, kết hợp cả hai thành phần tự hồi quy và trung bình di động. Mục đích của mô hình ARMA là cung cấp một mô hình **toàn diện hơn** cho chuỗi thời gian bằng cách **kết hợp cả tác động ngắn hạn và dài hạn**.

   * **Đặc điểm**: Cả ACF và PACF của mô hình ARMA đều giảm dần.
   * **Ứng dụng**: Mô hình ARMA được sử dụng rộng rãi trong phân tích chuỗi thời gian và có thể được áp dụng cho nhiều loại dữ liệu khác nhau.

**Mục đích chung của việc sử dụng các mô hình này trong phân tích chuỗi thời gian là:**

* **Mô hình hóa**: Mô tả mối quan hệ giữa các quan sát trong chuỗi thời gian.
* **Dự báo**: Dự đoán các giá trị tương lai của chuỗi thời gian dựa trên các mẫu trong quá khứ.
* **Kiểm soát**: Điều chỉnh một chuỗi thời gian bằng cách thao tác các biến đầu vào của nó dựa trên sự hiểu biết về động lực của chuỗi.

## Quy trình Tuyến Tính Tổng Quát (GLM)

Từ đây sẽ mở rộng ra 3 model MA, AR, ARMA.

### Giới thiệu tổng quan

**Quy trình tuyến tính tổng quát (Generalized Linear Process - GLP)** là một khung toán học tổng quát để mô hình hóa các quá trình tuyến tính phụ thuộc vào các giá trị quá khứ và nhiễu trắng (white noise). Đây là một công cụ quan trọng trong phân tích chuỗi thời gian, với các ứng dụng rộng rãi trong nhiều lĩnh vực. GLP bao gồm các mô hình phổ biến như Mô hình Trung bình Trượt (MA), Mô hình Tự Hồi Quy (AR), và Mô hình ARMA (AutoRegressive Moving Average).

### Định nghĩa và Công thức

Một quy trình tuyến tính tổng quát $\{Y_t\}$ được định nghĩa là một tổ hợp tuyến tính có trọng số của nhiễu trắng $\{ \epsilon_t \}$ hiện tại và quá khứ:

$$Y_t = \mu + \epsilon_t + \psi_1 \epsilon_{t-1} + \psi_2 \epsilon_{t-2} + \dots$$

Trong đó:
* $Y_t$: Giá trị của chuỗi thời gian tại thời điểm $t$.
* $\mu$: Trung bình của chuỗi, có thể khác 0.
* $\epsilon_t$: Nhiễu trắng tại thời điểm $t$, có kỳ vọng bằng 0 và phương sai $\sigma^2$.
* $\psi_i$: Các hệ số trọng số thể hiện mức độ ảnh hưởng của nhiễu trắng tại thời điểm $t-i$ đến giá trị của chuỗi tại thời điểm $t$.

### Điều kiện hội tụ

Để chuỗi $\{Y_t\}$ tồn tại và có các tính chất thống kê tốt, các hệ số trọng số $\psi_i$ cần thỏa mãn điều kiện hội tụ:

$$\sum_{i=1}^{\infty} |\psi_i| < \infty$$

Điều kiện này đảm bảo rằng ảnh hưởng của các nhiễu trắng trong quá khứ dần dần suy giảm và chuỗi $Y_t$ hội tụ về một giá trị trung bình hữu hạn.

### Tính chất của GLP

#### Trung bình:
Kỳ vọng của chuỗi $\{Y_t\}$ có thể được tính như sau:

$$E(Y_t) = \mu$$

#### Phương sai:
Phương sai của chuỗi $\{Y_t\}$ được xác định bởi phương sai của nhiễu trắng $\epsilon_t$ và các hệ số trọng số $\psi_i$:

$$\text{Var}(Y_t) = \sigma^2 \left( 1 + \sum_{i=1}^{\infty} \psi_i^2 \right)$$

#### Độ tự tương quan:
Độ tự tương quan (autocorrelation) tại độ trễ (lag) $k$ được biểu diễn bằng hiệp phương sai giữa $Y_t$ và $Y_{t-k}$:

$$\gamma(k) = \text{Cov}(Y_t, Y_{t-k})$$

Độ tự tương quan của GLP phụ thuộc vào các hệ số trọng số $\psi_i$ và sẽ giảm dần khi $k$ tăng.

#### Tính dừng:
GLP là một quá trình dừng (stationary) nếu kỳ vọng, phương sai và độ tự tương quan không thay đổi theo thời gian. Điều kiện hội tụ $\sum_{i=1}^{\infty} |\psi_i| < \infty$ là đủ để đảm bảo tính dừng của chuỗi $\{Y_t\}$.

#### Tính đảo ngược:
Một GLP được gọi là đảo ngược (invertible) nếu có thể biểu diễn nhiễu trắng $\epsilon_t$ dưới dạng một tổ hợp tuyến tính vô hạn của các giá trị hiện tại và quá khứ của chuỗi $Y_t$:

$$\epsilon_t = Y_t - \sum_{i=1}^{\infty} \phi_i Y_{t-i}$$

Tính đảo ngược là một thuộc tính quan trọng để ước lượng và dự báo trong các mô hình ARMA.

### Các mô hình đặc biệt của GLP

#### Mô hình MA (Moving Average):
Trong mô hình MA, giá trị hiện tại của chuỗi là tổ hợp tuyến tính của một số hữu hạn các nhiễu trắng quá khứ:

$$Y_t = \mu + \epsilon_t + \sum_{i=1}^{q} \psi_i \epsilon_{t-i}$$

#### Mô hình AR (AutoRegressive):
Trong mô hình AR, giá trị hiện tại của chuỗi là tổ hợp tuyến tính của các giá trị quá khứ của chính nó và nhiễu trắng hiện tại:

$$Y_t = \mu + \sum_{i=1}^{p} \phi_i Y_{t-i} + \epsilon_t$$

#### Mô hình ARMA (AutoRegressive Moving Average):
Mô hình ARMA kết hợp cả hai thành phần AR và MA:

$$Y_t = \mu + \sum_{i=1}^{p} \phi_i Y_{t-i} + \epsilon_t + \sum_{j=1}^{q} \psi_j \epsilon_{t-j}$$

Trong đó, $p$ là bậc của phần AR và $q$ là bậc của phần MA.

### Ứng dụng

GLP được sử dụng rộng rãi trong nhiều lĩnh vực, bao gồm:
* **Tài chính:** Dự báo giá chứng khoán, phân tích chuỗi thời gian giá cả và lãi suất.
* **Kinh tế:** Phân tích chuỗi thời gian kinh tế vĩ mô, mô hình hóa tiêu dùng và sản xuất.
* **Khoa học tự nhiên:** Phân tích tín hiệu trong các hệ thống vật lý và sinh học.
* **Xử lý ngôn ngữ tự nhiên:** Mô hình hóa các chuỗi văn bản hoặc âm thanh trong ngôn ngữ.

### Kết luận

Quy trình tuyến tính tổng quát (GLP) là một công cụ mạnh mẽ để mô hình hóa và phân tích chuỗi thời gian. Hiểu rõ các tính chất và ứng dụng của GLP là rất quan trọng để phát triển các mô hình dự báo chính xác và đáng tin cậy.

**Lưu ý:** Phân tích này chỉ mang tính tổng quan. Để hiểu sâu hơn về các khía cạnh toán học và phương pháp thống kê của GLP, cần tham khảo thêm các tài liệu chuyên ngành về chuỗi thời gian và lý thuyết xác suất.
